```python
# Reference: python graph gallery
# matplotlib documentation

# import libraries
from matplotlib import pyplot as plt
import numpy as np
#install matplot_venn using pip
from matplotlib_venn import venn3, venn3_circles
```

```python
# making a venn, reference: matplotlib documentation 
v = venn3(subsets = (1 ,1 ,1 ,1 ,1,1,1),set_labels = ('A','B','C'))

# custom venns 
v.get_patch_by_id('100').set_alpha(1.0)
#v.get_patch_by_id('100').set_color('white')
v.get_label_by_id('A').set_text('GOT S08E03 was cool')
v.get_label_by_id('B').set_text('Cool action and CGI')
v.get_label_by_id('C').set_text('People who are stupid')
c = venn3_circles(subsets = (1,1,1,1,1,1,1),linestyle = 'solid')
c[0].set_lw(3.0)
c[0].set_ls('dotted')


## title and arrow and text
plt.title("Game of thrones audience")
plt.annotate('You!', xy=v.get_label_by_id('111').get_position() - np.array([0, 0.05]), xytext=(-95,-70),
ha='center', textcoords='offset points', bbox=dict(boxstyle='round,pad=0.5', fc='gray', alpha=0.5),
arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.3',color='gray'))

plt.show()
plt.savefig('pie_charts.jpg', format='eps', dpi=1000)
```
